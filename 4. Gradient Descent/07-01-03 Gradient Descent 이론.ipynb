Gradient Descent 는 주어진 함수에 대해서 입력이 있을때 입력으로 함수를 미분하면은 우리는 그 함수의 출력값을 최소로 만드는 입력을 찾을수 있다. 
우리는 우리 머리속에 있는 함수를 모사하고 싶은데 우리가 가지고있는 Linear Layer나 Neural Network함수가 내뱉는 출력값과 실제값의 차이를 Loss이고 이 값을 통해 Loss function을 만들수 있다. 이 Loss function은 파라미터 입력을 받으며 함수는 파라미터에 의해 정의가 된다. 따라서 우리의 목적은 그 Loss function을 최소화하는 파라미터를 찾는게 목적이다. 따라서 gradient descent를 통해서 Loss function을 최소화로 만드는 입력 파라미터를 gradient  descent를 찾을수있고 그 과정에 미분이 필요하다.
