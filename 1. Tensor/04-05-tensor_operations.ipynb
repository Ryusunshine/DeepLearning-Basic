## Inplace Operations

```python
print(a)
print(a.mul(b)) #a 곱하기 b
print(a)
print(a.mul_(b)) #a의 자리에다가 a곱하기b 결과값을 덮어씌움. call by reference
print(a)
```

## Sum, Mean (Dimension Reducing Operations)

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]]) #하나의 스칼라값으로 결과값으로 나옴
```

```python
print(x.sum())
print(x.mean())
```

```python
print(x.sum(dim=0)) #특정 차원에서만 연산을 하고싶을때 dim = 을 설정해준다. 
#첫번째 dimension 값은 위에서 아래로, 두번째 dimension 값은 왼쪽에서 오른쪽으로 
print(x.sum(dim=-1))
```

## Broadcast in Operations 
= 모양이 다른 텐서들끼리 연산

What we did before,

```python
x = torch.FloatTensor([[1, 2]])
y = torch.FloatTensor([[4, 8]])
y
```

```python
z = x + y
print(z)
print(z.size())
```

Broadcast feature provides operations between different shape of tensors.

### Tensor + Scalar

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]])
y = 1

print(x.size())
```

```python
z = x + y
print(z)
print(z.size())
```

### Tensor + Vector

```python
x = torch.FloatTensor([[1, 2],
                       [4, 8]])
y = torch.FloatTensor([3,
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y 
print(z)
print(z.size())
Image("C:\\Users\\ryusu\\OneDrive\\바탕 화면\\새 폴더\\2.png")
```

```python
x = torch.FloatTensor([[[1, 2]]]) #3차원. ([[[1,1,2]]])랑 똑같
y = torch.FloatTensor([3,
                       5]) #벡터([2])라서 애를 3차원으로 만들어준다.([[[1,1,2]]]). 앞에 1,1를 붙임

print(x.size())
print(y.size()) 
```

```python
z = x + y
print(z) #사이즈를 같게해주고 값끼리 연산.
print(z.size()) #사이즈는 3차원 ([1,1,2])
Image("C:\\Users\\ryusu\\OneDrive\\바탕 화면\\새 폴더\\3.png")
```

### Tensor + Tensor

```python
x = torch.FloatTensor([[1, 2]])
y = torch.FloatTensor([[3],
                       [5]])

print(x.size())
print(y.size()) #두개의 사이즈가 다름.그래서 사이즈를 같게해주려 값들을 복사함.
```

```python
z = x + y
print(z)
print(z.size())
```

x, y 가 밑으로 옆으로 하나씩 늘어나서 같은 사이즈의 행렬을 만들고 연산을 한다

### Failure Case

```python
x = torch.FloatTensor([[[1, 2],
                        [4, 8]]])
y = torch.FloatTensor([[1, 2, 3],
                       [4, 5, 6],
                       [7, 8, 9]])

print(x.size())
print(y.size())

z = x + y
```
