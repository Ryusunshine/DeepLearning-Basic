```python
a * b
```

```python
a / b
```

```python
a == b
```

```python
a != b
```

## Inplace Operations

```python
print(a)
print(a.mul(b)) #a  b
print(a)
print(a.mul_(b)) #a  ab  . call by reference
print(a)
```

## Sum, Mean (Dimension Reducing Operations)

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]]) #   
```

```python
print(x.sum())
print(x.mean())
```

```python
print(x.sum(dim=0)) #    dim =  . 
# dimension   ,  dimension    
print(x.sum(dim=-1))
```

## Broadcast in Operations

What we did before,

```python
x = torch.FloatTensor([[1, 2]])
y = torch.FloatTensor([[4, 8]])
y
```

Broadcast feature provides operations between different shape of tensors.

### Tensor + Scalar

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]])
y = 1

print(x.size())
```

```python
z = x + y
print(z)
print(z.size())
```

### Tensor + Vector

```python
x = torch.FloatTensor([[1, 2],
                       [4, 8]])
y = torch.FloatTensor([3,
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y 
print(z)
print(z.size())
Image("C:\\Users\\ryusu\\OneDrive\\ \\ \\2.png")
```

```python
x = torch.FloatTensor([[[1, 2]]]) #3. ([[[1,1,2]]]) 
y = torch.FloatTensor([3,
                       5]) #([2])  3 .([[[1,1,2]]]).  1,1 

print(x.size())
print(y.size()) 
```

```python
z = x + y
print(z) #   .
print(z.size()) # 3 ([1,1,2])
Image("C:\\Users\\ryusu\\OneDrive\\ \\ \\3.png")
```

### Tensor + Tensor

```python
x = torch.FloatTensor([[1, 2]])
y = torch.FloatTensor([[3],
                       [5]])

print(x.size())
print(y.size()) #  .    .
```

```python
z = x + y
print(z)
print(z.size())
```
