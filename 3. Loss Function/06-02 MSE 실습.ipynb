# Mean Square Error (MSE) Loss

```python
import torch
from IPython.display import Image
```

$$
\text{MSE}(\hat{x}_{1:N}, x_{1:N})=\frac{1}{N\times{n}}\sum_{i=1}^{N}{||x_i-\hat{x}_i||_2^2}
$$

```python
def mse(x_hat, x):
    # |x_hat| = (batch_size, dim)
    # |x| = (batch_size, dim)
    y = ((x - x_hat)**2).mean() #그림에서 가로로 파란색선, 위로향하는 파란색선 곱한거의 평균값구한값임 
    
    return y

Image("C:\\Users\\ryusu\\OneDrive\\바탕 화면\\8.png")
```

```python
x = torch.FloatTensor([[1, 1],
                       [2, 2]])
x_hat = torch.FloatTensor([[0, 0],
                           [0, 0]])

print(x.size(), x_hat.size())
```

```python
mse(x_hat, x)
```
