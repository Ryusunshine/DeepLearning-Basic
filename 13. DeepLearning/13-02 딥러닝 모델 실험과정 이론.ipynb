= 그래서 학습을 진행하면서 train set의 loss랑 validation set의 loss를 비교하면서 둘이 같이 내려가고있는지 확인하면 된다. 
* validation set의 loss가 generalization erro(loss)임

* Training(학습)
iteration 시작 -> feed-forward(미니배치를 해주고 모델에 넣고 돌리기) -> y_hat 이 나오면 y와 비교해서 loss 계산(loss = F.mse_loss(y_hat_i, y_i)) - > loss가 나오면 미분해서 back-propagation 진행(loss.backward()) - > gradient 구해졌으니깐 gradient descent 수행 (optimizer.step() ) -> 사용자의 현재상태를 출력 -> 그럼 하나의 미니배치(mini-batch)의 iteration 이 끝남 ( 데이터가 N개있을때 배치사이즈 K에 대해서 N/K번 만큼 iteration을 진행한다) = 한 epoch이 끝남


* Validation(평가)
validation 도 마찬가지로 mini-batch로 나눠서 간다. 미니배치를 feed-forward를 하면 y_hat이 나옴 -> y_hat과 y비교해서 loss를 구한다 -> 사용자에게 validation set의 loss를 보여줌 -> iteration 종료. 이 순서를 반복 -> best loss 여부 체크(validation loss중에 좋았던 애들 체크(예전보다 validation set의 loss가 낮아졌니?)-> 모델 저장 -> 만약 다음 epoch의 validation loss가 전의 epoch의 ㅣloss 보다 낮아지지 않았다면 모델을 저장하지않는다) -> 정해진 epoch의 횟수만큼 돈다
